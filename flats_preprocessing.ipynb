# Cleaning Flats data


```python
flats.head()
```

```python
flats.shape
```

```python
flats.info()
```

```python
## dropping unnecessary columns

flats.drop(['link','property_id'],axis =1,inplace = True)
```

```python
## renaming column

flats.rename(columns= {'area' :'price_per_sqft'}, inplace= True)
```

```python
flats.head()
```

```python
flats['society'].value_counts().shape
```

```python
flats['society'] = flats['society'].apply(lambda name : re.sub(r'\d+(\. \d+)?\s★','', str(name)).strip() ).str.lower()
```

```python
flats['society'].value_counts().shape
```

```python
flats['price'].value_counts()
```

```python
## Cleaning price feature and scaling it to crore range

def treat_price(x):
    if isinstance(x, float):
        return x
    else:
        parts = x.split(' ')
        if len(parts) == 2 and parts[1] == 'Lac':
            return round(float(parts[0])/100, 2)
        elif len(parts) == 2 and parts[1] == 'Crore':
            return round(float(parts[0]), 2)
        else:
            return None  # Handle unexpected cases
```

```python
flats = flats[flats['price'] != 'Price on Request']
```

```python
flats['price'] = flats['price'].apply(treat_price)
```

```python
flats.head()
```

```python
## Cleaning price_per_sqft

flats['price_per_sqft'] = flats['price_per_sqft'].replace('[^\d]','',regex= True) \
                               .replace('', np.nan) \
                               .astype(float)
```

```python
flats = flats[~flats['bedRoom'].isnull()]
flats = flats[~flats['bathroom'].isnull()]
flats = flats[~flats['balcony'].isnull()]
```

```python
## bedrooms

flats['bedRoom'] = flats['bedRoom'].replace('[^\d]','',regex= True) \
                               .replace('', np.nan) \
                               .astype(int)
```

```python
## bathroom

flats['bathroom'] = flats['bathroom'].replace('[^\d]','',regex= True) \
                               .replace('', np.nan) \
                               .astype(float).astype(int)
```

```python
flats['balcony'].value_counts()
```

```python
## balcony

flats['balcony'] = flats['balcony'].str.split(' ').str.get(0).str.replace('No','0')
```

```python
flats.head()
```

```python
flats['additionalRoom'].value_counts()
```

```python
flats['additionalRoom'].fillna('None', inplace= True)
```

```python
flats['floorNum'] = flats['floorNum'].str.split(' ').str.get(0) \
                              .replace('Ground','0').str.replace('Basement','-1')\
                              .str.replace('Lower','0').str.extract(r'(\d+)')\
                              .astype(float)
```

```python
flats['facing'].isnull().sum()
```

```python
flats['facing'].fillna('None', inplace=True)
```

```python
flats.insert(4,'area',round((flats['price']*1000000)/flats['price_per_sqft']))
```

```python
flats['area'] = flats['area'] * 10
```

```python
flats.insert(1,'property_type','flat')
```

```python
flats.to_csv('flats_cleaned.csv', index= False)
```
