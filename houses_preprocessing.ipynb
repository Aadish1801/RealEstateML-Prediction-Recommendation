```python
import pandas as pd
import numpy as np
import re
```

```python
df = pd.read_csv('independent_house.csv')
```

```python
pd.set_option('display.max_columns',None)
pd.set_option( 'display.max_rows',None)
```

```python
df.drop(['link','property_id'], axis= 1 , inplace= True)
```

```python
df['price'].value_counts()
```

```python
df = df[df['price'] != 'Price on Request']
```

```python
def treat_price(x):
    if isinstance(x, float):
        return x
    else:
        parts = x.split(' ')
        if len(parts) == 2 and parts[1] == 'Lac':
            return round(float(parts[0])/100, 2)
        elif len(parts) == 2 and parts[1] == 'Crore':
            return round(float(parts[0]), 2)
        else:
            return None  # Handle unexpected cases
```

```python
df['price'] = df['price'].apply(treat_price)
```

```python
df.head()
```

```python
df['rate'].isnull().sum()
```

```python
df['rate'].value_counts()
```

```python
df['rate'] = (df['rate'].replace('[^\d]', '',regex= True)
                       .replace('', np.nan)
                       .astype(float)   )
```

```python
df.head()
```

```python
df.dropna(subset=['bedRoom', 'bathroom', 'balcony'],inplace=True)
```

```python
## bedrooms

df['bedRoom'] = df['bedRoom'].replace('[^\d]','',regex= True) \
                               .replace('', np.nan) \
                               .astype(int)
```

```python
## bathroom

df['bathroom'] = df['bathroom'].replace('[^\d]','',regex= True) \
                               .replace('', np.nan) \
                               .astype(float).astype(int)
```

```python
df['balcony'].value_counts()
```

```python
## balcony

df['balcony'] = df['balcony'].str.split(' ').str.get(0).str.replace('No','0')
```

```python
df['balcony'].value_counts()
```

```python
df.head()
```

```python
df['additionalRoom'].fillna('None', inplace= True)
```

```python
df['additionalRoom'] = df['additionalRoom'].str.lower()
```

```python
df['noOfFloor'] = df['noOfFloor'].replace('[^\d]','',regex= True) \
                               .replace('', np.nan) \
                               .astype(float)
```

```python
df['noOfFloor'].isnull().sum()
```

```python
df.rename(columns={'noOfFloor':'floorNum'},inplace=True)
```

```python
df['facing'].fillna('NA',inplace=True)
```

```python
df.insert(loc=1,column='property_type',value='house')
```

```python
df['society'] = df['society'].apply(lambda name: re.sub(r'\d+(\.\d+)?\s?★', '', str(name)).strip()).str.lower()
```

```python
df['additional']= df['additionalRoom'].str.lower()
```

```python
df['society'] = df['society'].str.replace('nan','independent')
```

```python
df.head()
```

```python
df.rename(columns={'rate':'price_per_sqft'},inplace=True)
```

```python
df.head()
```

```python
df.to_csv('house_cleaned.csv',index=False)
```
